```python
## For loop
for i in range(5):
    print(i)
```

```python
## Break Statement
list1 = [10, 20, 30, 40,50, 60]
for i in list1[:4]:
    if i == 40:
        break
    print(i)
```

```python
## Continue Statement
list2 = [10, 30, 50, 70, 90]
for i in list2[:5]:
    if i == 50:
        continue
    print(i)
```

```python
## If, Nested if, Else statement
age = 40
if age == 30:
    print("You are an adult.")
    if age>= 19:
        print("You are teenager.")
else:
    print("You are senior adult.")
        
```
