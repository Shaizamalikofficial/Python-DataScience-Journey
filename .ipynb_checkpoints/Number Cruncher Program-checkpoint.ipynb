### Task3: Number Cruncher Program

```python
# Use a for loop from 1 to 20
for num in range(1, 21):

    # Apply conditions
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")           # divisible by both 3 and 5
    elif num % 3 == 0:
        print("Fizz")              # divisible by 3
    elif num % 5 == 0:
        print("Buzz")             # divisible by 5
    else:
        print(num)               # otherwise print the number

# Ask user for a number
user_num = int(input("\nEnter a number for countdown: "))

# If number is greater than 10, stop using break
if user_num > 10:
    print("Number is greater than 10. Countdown stopped!")
else:
    # Use while loop for countdown
    while user_num >= 1:
        print("Countdown:", user_num)
        user_num -= 1        # decrease number by 1 each time

    print("Blast Off!")
 
```
