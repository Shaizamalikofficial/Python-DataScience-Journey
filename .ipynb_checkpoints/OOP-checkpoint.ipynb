```python
## Input animal from user and check if animal exist then print its sound.
class AnimalSounds:
    def __init__(self):
        self.animals = {"Dog": "Wauuu Wauu", "Cat": "Meow", "Lion": "Roar", "Sheep": "Beehh", "Wolf": "Aaouu", "Donkey": "Dheachuu Dheachuu"}
    def check_animal(self):
        user_animal = input("Enter animal name:")
        if user_animal in self.animals:
            print("Sound:", self.animals[user_animal])
        else:
            print("The animal is not available in list.")
obj = AnimalSounds()
obj.check_animal() 
```

```python
## Input animal from user and check if exist or not. 
class AnimalChecking:
    def __init__(self):
        self.animals = ['Dog', 'Cat', 'Lion', 'Tiger', 'Goat', 'Sheep', 'Snake', 'Zebra', 'Giraffa']
    def check_animal(self):
        user_animal = input("Enter Animal Name:")
        if user_animal in self.animals:
            print("Animal exists in list")
        else:
            print("Animal does not exist")
obj = AnimalChecking()
obj.check_animal()
```

```python
class Mobile:
    def __init__(self):
        self.mobiles = ['Nokia', 'Techno', 'Realme', 'Oppo', 'Infinix', 'Google', 'Samsung', 'IPhone']
    def check_mobile(self):
        user_mobile = input('Enter your mobile phone type:')
        if user_mobile in self.mobiles:
            print("Mobile Phone exists in list")
        else:
            print("Mobile Phone does not exist")
    def add_item (self):
        New_phone =input("Enter new mobile type:")
        self.mobiles.append(New_phone)
        print ("Updated List:", self.mobiles)
        
obj = Mobile()
obj.check_mobile()
obj.add_item()
```
