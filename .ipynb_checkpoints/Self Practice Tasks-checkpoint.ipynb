```python
## Task 12: Check age >=18 and citizen
age = int(input("Enter your age:"))
citizen = input("Are you a citizen(yes/no):")
if age>=18:
    print("age is valid")
    if citizen == "yes":
        print("You are eligible to vote")
    else:
        print("not eligible to vote (not a citizen)")
else:
    print("not eligible (age is less than 18)")
    
```

```python
## Task 13: Classify marks (pass & grade)
marks = int(input("Enter your marks:"))
if marks >= 50:
    if marks>=90:
        print("You have passed.Your Grade is A+")
    elif marks>=80:
        print("You have passed. Your Grade is A")
    elif marks >= 70:
        print("You have passed.Your Grade is B")
    elif marks >= 60:
        print("You have passed.Your Grade is C")
    else:
        print("You have passed.Your Grade is D")
else:
    print("You have failed with Grade F")
    
```

```python
## Task 15: Password + OTP check simulation
correct_password = 'abc1234'
correct_otp = 2468

password = input ("Enter your password:")
if password == correct_password:
    otp = int(input("Enter your otp:"))
    if otp == correct_otp:
        print("Login Successfully")
    else:
        print("Incorrect otp")
else:
    print("Incorrect Password")
    
```

```python
## Task 16: Salary > threshold then tax bracket
salary = int(input("Enter your salary:"))
threshold = 50000
if salary > threshold:
    print("You have to pay tax")
    if salary <=100000:
        print("Tax bracket: 10%")
    elif salary<=200000:
        print("Tax bracket: 20%")
    else:
        print("Tax bracket: 30%")
else:
    print("No Tax")
```
