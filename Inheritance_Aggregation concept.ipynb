## Inheritance  concept, Aggregation concept

```python
## Aik class ky data ko bahir access krna without inheritance called Aggregation

# Parent class
class Animal:
    def __init__(self, name = 'Cat', color= 'Red'):
        self.name = name            # attribute 1
        self.color = color           # attritube 2

    def Info(self):                  # method
         print("name is:", self.name, "Color is", self.color)
# Object
a = Animal()
a.Info()       ## Aggregation 
```

```python
# Child Class Dog Inherits Animal
class Dog(Animal):
    def __init__(self, name, color, age, no, loyal):
        super().__init__(name, color)               # Parent constructor cell
        self.age = age
        self.no = no
        self.loyal = loyal
    
    def Run(self):
        if self.age>0:
            print("Animal:", self.name, "can run")
        else:
            print("Animal:", self.name, "cannot run")

# Object create
d = Dog("German sheferd", "Dark Brown", 2, 6, True)
d.Run()
```

```python
class Employee:
    def __init__(self, name, employee_id, salary):
        self.name = name
        self.employee_id = employee_id
        self.salary = salary
        self._department = "General"         # Protected attribute

# Protected Method:
    def set_department(self, depart):
        self._department = depart

# Display method
    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Employee ID: {self.employee_id}")
        print(f"Salary: {self.salary}")
        print(f"Department: {self._department}")

class Manager(Employee):
    def __init__(self, name, employee_id, salary, team_size):
## Call parent class constructor
       super().__init__(name, employee_id, salary)
       self.__team_size = team_size           # Private attribute
       self.set_department("Management")      # Using parent's protected method

# Additional methods specific to Manager
# Getter Method
    def get_team_size(self):
        return self.__team_size
# Setter Method
    def set_team_size(self, new_size):
        if new_size > 0:
            self.__team_size = new_size
            return True
        else:    
            return False

## Override parent method(polymorphism)
## def calculate_bonus(self):
#     """Override: Managers get 10% bonus + extra per team member"""
#     base_bonus = self.__team_size*500
#     return base_bonus + team_bonus

# Override Display Method
    def display_info(self):
        super().display_info()          # Call parent display
        print(f"Team Size: {self.__team_size} members")
        print(f"Management Bonus: +${self.__team_size*500:,.2f}")

# Employee object
e1 = Employee("Ali", 101, 40000)
e1.display_info()

print("\n")

# Manager object
m1 = Manager("Sara", 102, 60000, 5)
m1.display_info()
```
