### Polymorphism Concept

```python
# Create a base class Animal with method sound().Create subclasses: • Dog • Cat • Cow. Each overrides sound() and prints a different sound. 

## Base class
class Animal:
    def sound(self):
        print("Animal makes different sounds")

## Subclass Dog
class Dog(Animal):
    def sound(self):
        print("Dogs Bark")

## Subclass Cat
class Cat(Animal):
    def sound(self):
        print("Cat Meao")

## Subclass Cow
class Cow(Animal):
    def sound(self):
        print("Cows Mooos")

## Objects
d = Dog()
c = Cat()
C = Cow()

## Sound Calling
d.sound()
c.sound()
C.sound()
```

```python
# Create a base class Vehicle with method move(). Subclasses: • Car • Bike • Bus .Store objects in a list and call move() using a 
# loop to demonstrate runtime polymorphism.

## Base class
class Vehicle:
    def move(self):
        print("Vehicle is moving")

## Subclass Car
class Car(Vehicle):
    def move(self):
        print("Car is driving")

## Subclass Bike
class Bike(Vehicle):
    def move(self):
        print("Bike is running")

## Subclass Bus
class Bus(Vehicle):
    def move(self):
        print("Bus is transporting passengers")

## Objects list
vehicles = [Car(), Bike(), Bus()]

## Loop for polymorphism
for v in vehicles:
    v.move()
```

```python
# Create base class Shape with method area(). Subclasses: • Rectangle (length × width) • Circle (πr²) • Triangle (½ × base × height) 
# Override area() in each class.

## Base class
class Shape:
    def area(self):
        print("Area")

## Subclass Rectangle
class Rectangle(Shape):
    def area(self):
        print("Area of Rectangle = length * width")

## Subclass Circle
class Circle(Shape):
    def area(self):
        print("Area of Circle = πr²")

## Subclass Triangle
class Triangle(Shape):
    def area(self):
        print("Area of Triangle = ½ * base * height")

## Objects
r = Rectangle()
c = Circle()
t = Triangle()

## Area calling
r.area()
c.area()
t.area()
```

```python
# Base class Employee with method get_role(). Subclasses: • Manager • Developer • Designer 
# Each should return a different role description. 

## Base Class
class Employee:
    def get_role(self):
        return("Employee Level")

## Subclass Manager
class Manager(Employee):
    def get_role(self):
        return("Manager manages the team")

## Subclass Developer
class Developer(Employee):
    def get_role(self):
        return("Developer writes code")

## Subclass Designer
class Designer(Employee):
    def get_role(self):
        return("Designer creates UI designs")

## Objects
m = Manager()
d = Developer()
b = Designer()

## Role Calling
print(m.get_role())
print(d.get_role())
print(b.get_role())
```

```python
# Payment Processing System Base class Payment with method pay(amount). Subclasses: • CreditCardPayment • JazzCashPayment • BankTransferPayment 
# Each should print a different transaction message. 

## Base class
class Payment:
    def pay(self, amount):
        print("Processing payment of", amount)

# Subclass Credit Card
class CreditCardPayment(Payment):
    def pay(self, amount):
        print("Payment of", amount, "done using Credit Card")


# Subclass JazzCash
class JazzCashPayment(Payment):
    def pay(self, amount):
        print("Payment of", amount, "done using JazzCash")


# Subclass Bank Transfer
class BankTransferPayment(Payment):
    def pay(self, amount):
        print("Payment of", amount, "done using Bank Transfer")


# Objects list
payments = [CreditCardPayment(), JazzCashPayment(), BankTransferPayment()]

# Loop (Runtime Polymorphism)
for p in payments:
    p.pay(5000)
```
