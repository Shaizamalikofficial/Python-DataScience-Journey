```python
class Student1:
    def __init__(self):
        self.__name = "Ali"   # private variable
        self.age = 20

    def display(self):
        print("Name:", self.__name)
        print("Age:", self.age)
# Getter        
    def get_name(self):
        return self.__name
## Setter
    def set_name(self, name):
        self.__name = name


# object 
s1 = Student1()

print("First time")
s1.display()

print("Try to change directly")
s1.__name = "Ahmed"     # this will not change original private variable
print(s1.get_name())    # still Ali

print("Again change using setter")
s1.set_name("Ahmed")    # correct way
s1.display()
```

```python
class Employee:
    def __init__(self):
        self.__name = "Shaiza"
        self.__role = "Python Developer"
        self.__salary = 50000
    def display(self):
        print("Name:", self.__name)
        print("Role:", self.__role)
        print("Salary:", self.__salary)
# Getter
    def get_name(self):
        return self.__name

    def get_role(self):
        return self.__role

    def get_salary(self):
        return self.__salary
#Setter
    def set_name(self, new_name):
        self.__name= new_name

    def set_role(self, new_role):
        self.__role= new_role

    def set_salary(self, new_salary):
        self.__salary = new_salary

# Object
emp1 = Employee()

print("First time")

print(emp1.get_name())
print(emp1.get_role())
print(emp1.get_salary())

print("After changing")

emp1.set_name("Sara")
emp1.set_role("Manager")
emp1.set_salary(70000)

print(emp1.get_name())
print(emp1.get_role())
print(emp1.get_salary())
```

```python
## Employee Bonus Calculation:
class Employee:
    def __init__(self, name, role, salary):
        self.__name = name
        self.__role = role
        self.__salary = salary
## Getter:
    def get_name(self):
        return self.__name

    def get_role(self):
        return self.__role

    def get_salary(self):
        return self.__salary

## Bonus Method:
    def calculate_bonus(self):
        role = self.get_role()
        salary = self.get_salary()
        
        if role == "Manager":
            bonus = salary * 0.25
        elif role == "Developer":
            bonus = salary * 0.15
        else:
            bonus = salary * 0.5
        print ("Bonus:", bonus)

## Object:
obj = Employee('Shaiza', 'Manager', 70000)
print("Name:", obj.get_name())
print("Role:", obj.get_role())
print("Salary:", obj.get_salary())
obj.calculate_bonus()
```

```python
## Student Grading System by using class methods & object:
class Students:
    def __init__(self):
        self.names = ['Ayesha', 'Aslam', 'Arsalan', 'Urwa', 'Ayat', 'Ibrahim', 'Adan', 'Minal', 'Zainab', 'Zaisha']
        self.marks = [90, 95, 86, 88, 79, 99, 67, 59, 49, 77]

    def grade(self, marks):
        if marks >=90:
            return "A+"
        elif marks>=80:
            return "A"
        elif marks>=70:
            return "B"
        elif marks>=60:
            return "C"
        elif marks >=50:
            return "D"
        else:
            return "Fail"

    def check_students(self):
        name = input("Enter your Name:")
        if name in self.names:
            index = self.names.index(name)
            marks = self.marks[index]
            grade = self.grade(marks)

            print("Name:", name)
            print("Marks:", marks)
            print("Grade:", grade)
        else:
            print("Student Not Found")

obj = Students()
obj.check_students()
```
